```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_English-speaking_population'
```

```python
table = pd.read_html(url)
table
```

```python
df_full_table = table[1]
df_full_table.head()
```

```python
df_short_table = df_full_table[["Country","Eligible population","Total English speakers"]]
df_short_table = df_short_table.drop(126)
df_short_table
```

```python
df_short_table["Total English speakers"] = df_short_table["Total English speakers"].astype(float)
df_short_table["Eligible population"] = df_short_table["Eligible population"].astype(float)
df_short_table['Percent_English_Speakers'] = df_short_table["Total English speakers"] / df_short_table["Eligible population"]
df_short_table
```

```python
df_short_table = df_short_table.rename(columns = {'Eligible population': 'Eligible_Population', 'Total English speakers': 'Total_English_Speakers'})
df_short_table
```

```python
url_2 = 'https://www.listofcountriesoftheworld.com/area-land.html'
```

```python
area_table = pd.read_html(url_2)
area_table
```

```python
df_full_area_table = area_table[0]
df_full_area_table.head()
```

```python
df_info_table = df_short_table.merge(df_full_area_table,on="Country",how="inner")
df_info_table = df_info_table.drop(columns="#")
df_info_table = df_info_table.rename(columns = {"Land area":"Land_Area"})
df_info_table
```

```python
df_info_table["Population_Density"] = df_info_table["Eligible_Population"] / df_info_table["Land_Area"]
df_info_table
```

```python
df_info_table.to_csv("Info_Table.csv")
```
