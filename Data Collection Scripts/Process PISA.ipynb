```python
import pandas as pd
import numpy as np

file_name = '72a5eb80-96fe-45b0-a583-4f545954bc73_Data.csv'
raw_df = pd.read_csv(file_name)
raw_df.head(1)
```

```python
#Remove extraneous data

clean_df = raw_df[["Country Name","Country Code","Series Name","2018 [YR2018]"]]
clean_df = clean_df.dropna()
clean_df = clean_df.rename(columns={"Country Name": "country_name","Country Code": "country_code","Series Name":"series","2018 [YR2018]":"score"})
clean_df = clean_df.drop(clean_df[clean_df["score"] == ".."].index)
clean_df = clean_df.drop(clean_df[(clean_df["series"] != "PISA: Mean performance on the mathematics scale") & 
                                  (clean_df["series"] != "PISA: Mean performance on the reading scale") & 
                                  (clean_df["series"] != "PISA: Mean performance on the science scale")].index)
clean_df = clean_df.drop(columns = "series")
clean_df
```

```python
# Group by country and average PISA score

clean_df["score"] = clean_df['score'].astype(float)
clean_df = clean_df.groupby(['country_name','country_code'])['score'].mean().reset_index()
clean_df
```

```python
clean_df.to_csv("PISA.csv")
```
